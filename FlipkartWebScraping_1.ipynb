```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib.parse
```

```python
website ="https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_7c0f7cf4-13e1-4dfb-9666-96a70f583f16_1_372UD5BXDFYS_MC.34WHNYFH5V2Y&otracker=hp_rich_navigation_8_1.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop_34WHNYFH5V2Y&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_8_L1_view-all&cid=34WHNYFH5V2Y"
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
results= soup.find_all('div', {'class':"_13oc-S"})
```

```python
relative_url = results[0].find('a', {'class':"_1fQZEK"}).get('href') #extract relative url
relative_url
```

```python
#root url
root_url = 'https://www.flipkart.com/'
```

```python
#combine root url and relative url
combined_url = root_url + relative_url
combined_url
```

```python
results[0].find('div', {'class':"fMghEO"}).get_text() # prod details
```

```python
#putting data points inside a for loop
product_name= []
product_price=[]
product_reviewStar= []
product_reviewCount= []
product_relativeUrl= []
product_Details= []


for result in results: 
    #name
    try:
        product_name.append(result.find('div', {'class': '_4rR01T'}).get_text())
    except:
        product_name.append("N/A")
    
    #price
    try:
        product_price.append(result.find('div', {'class': '_30jeq3 _1_WHN1'}).get_text().strip()[1:])
    except:
        product_price.append("N/A")
    
    #star rating
    try:
        product_reviewStar.append(result.find('div', {'class': '_3LWZlK'}).get_text())
    except:
        product_reviewStar.append("N/A")
    
    #review count
    try:
        product_reviewCount.append(result.find('span', {'class': '_2_R_DZ'}).get_text().strip()[:2])
    except:
        product_reviewCount.append("N/A")
    
    #relative url
    try:
        product_relativeUrl.append(result.find('a', {'class':"_1fQZEK"}).get('href'))
    except:
        product_relativeUrl.append("N/A")
        
    #product details
    try:
        product_Details.append(result.find('div', {'class':"fMghEO"}).get_text())
    except:
        product_Details.append("N/A")
    
```

```python
#combine the url

url_combined = []

for link in product_relativeUrl:
    url_combined.append(urllib.parse.urljoin(root_url, link))
```

```python
url_combined
```

```python
#creating pandas data frame
product_overview = pd.DataFrame({'Name' : product_name, 'Price' : product_price, 'StarRating': product_reviewStar, 
                                 'ReviewCount': product_reviewCount, 'Link': url_combined, 'ProductDetails' : product_Details})
```
